```python
df3 = df.copy()
```

```python
def length_imputer(series):
    if series.median() < series.mean():
        result = series.median()
    else:
        result = series.mean()
    return series.fillna(result)

df3['LENGTH'] = df3.groupby('RIVER')['LENGTH'].transform(length_imputer)
```

```python
df3["LENGTH"] = df3.groupby('RIVER').LENGTH.transform(lambda series : series.fillna(series.median() if series.median() < series.mean() else series.mean()))
```

```python
df3.head(10)
```

```python
df3.transform({
    'LENGTH' : lambda series : series.median(),
    'LANES' : lambda item : -1 if not item else item
}, inplace=True)

df3.head(10)
```

```python
df4 = df3.transform({
    'LENGTH':lambda series : series.fillna(series.median() if series.median() < series.mean() else series.mean()),
    'LANES': lambda item: -1 if pd.isna(item) else item
})

df4.head()
```
